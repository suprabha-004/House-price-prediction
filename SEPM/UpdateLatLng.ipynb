```python
import pandas as pd
import requests
import time

# Load your data
file_path = "C:/Users/ASUS/.cache/kagglehub/datasets/pradeepsapparapu/bengaluru-house-datacsv/versions/1/Bengaluru_House_Data.csv"
data = pd.read_csv(file_path)

# Function to get latitude and longitude from Nominatim
def get_coordinates(location):
    url = f"https://nominatim.openstreetmap.org/search?q={location},Bengaluru&format=json"
    response = requests.get(url)
    if response.status_code == 200:
        try:
            data = response.json()[0]
            return data['lat'], data['lon']
        except (IndexError, KeyError):
            return None, None
    else:
        return None, None

# Apply function to each unique location
data['latitude'] = None
data['longitude'] = None

for i, location in enumerate(data['location'].unique()):
    lat, lon = get_coordinates(location)
    data.loc[data['location'] == location, 'latitude'] = lat
    data.loc[data['location'] == location, 'longitude'] = lon
    print(f"Processed {location}: ({lat}, {lon})")
    time.sleep(1)  # To respect Nominatim's rate limit

# Save the updated data to a new CSV
data.to_csv('Bengaluru_House_Data_with_Coordinates.csv', index=False)
print("Coordinates added and saved to Bengaluru_House_Data_with_Coordinates.csv")
```
